```python
import pandas as pd
import ftfy 

cols = ['title', 'description', 'phone_model']
for col in cols:
    data[col] = data[col].apply(lambda x: ftfy.fix_text(str(x)) if pd.notna(x) else x)

data.head(164)
```

```python
location_words = ['cairo', 'giza', 'alexandria', 'nasr city', 'mokattam', 
                  'maadi', 'marg', 'heliopolis', 'zamalek', 'shorouk', 
                  'madinaty', 'sheikh zayed', 'corniche']
mask_location = data['title'].str.lower().apply(lambda x: any(w in x for w in location_words))
data = data[~mask_location]

# 3. Drop accessories (Amazon cases, earphones, etc.)
accessory_words = ['case', 'screen protector', 'tempered glass', 'earphone',
                   'headphone', 'cable', 'charger', 'cover', 'adapter']
mask_accessory = data['title'].str.lower().apply(lambda x: any(w in x for w in accessory_words))
data = data[~mask_accessory]

# 4. Drop duplicates
data = data.drop_duplicates(subset='id')

# 5. Reset index
data = data.reset_index(drop=True)

print(f"Total rows: {len(data)}")
print(f"Trusted (0): {len(data[data['label']==0])}")
print(f"Scammy  (1): {len(data[data['label']==1])}")
data.head()
```
